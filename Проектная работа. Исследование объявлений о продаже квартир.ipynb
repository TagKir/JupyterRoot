```python
import pandas as pd
from heapq import nlargest
import matplotlib.pyplot as plt
import datetime as dt   

# 1:

df = pd.read_csv(r"C:\Users\tagir\Downloads\real_estate_data.csv"  ,sep='	')

# 2 Задание, вставляю пропущ. знач. :

df['balcony'] = df['balcony'].fillna(0)
df['cityCenters_nearest'] = df['cityCenters_nearest'].fillna('SO Far')
df['airports_nearest'] = df['airports_nearest'].fillna('SO Far')


# 3.1:
df['Price_Per_Square'] = (df['last_price'] / df['living_area']).round()
# 3.2:
# Оно сделается в 4.2
# 3.3:
# if df['floor'] == df['floors_total']:
#     df.insert(loc='last', column='floor_variable', value=new_col)
df.insert(loc=9, column='floor_variable', value=None)
df.loc[df['floor'] == df['floors_total'], 'floor_variable'] = 'last' 
df.loc[df['floor'] <= 1, 'floor_variable'] = 'first' 
df['floor_variable'] = df['floor_variable'].fillna('other')
# УРАААА рабоотает
# 3.4:
df['corr_livAndTotal_area'] = (df['living_area'] / df['total_area']).round(3)
df['corr_kitAndTotal_area'] = (df['kitchen_area'] / df['total_area']).round(3)



# 4.1:
# df.hist(bins = 20, column='ceiling_height', range=(1, 5))
# plt.show()
# print(df['ceiling_height'].value_counts())
# 4.2:
df['diff'] = str(dt.date.today())

df['diff'] = pd.to_datetime(df['diff'], errors='coerce')
df['first_day_exposition'] = pd.to_datetime(df['first_day_exposition'], errors='coerce')

df['diff'] = (df['diff'] - df['first_day_exposition']).dt.total_seconds() / 86400 # дни
# df.hist(bins = 200, column = 'diff')
# plt.show()
print(round(df['diff'].mean())) # 1875
print(round(df['diff'].median())) # 1842
# Продажа хорошего(имеющего спрос) жилья занимает в среднем не более 2350 дней, дальше на гистограмме
# идёт резкое повышение. Это дома(квартиры), которым чего-либо не хватает для продажи.
# Если жильё продали менее чем за 1500 - рано.
# Если позже чем за 2350 - поздно



# 4.3
# last price - не выбивается
# total_area - не выбивается
# rooms - не выбивается
# ceiling_height - не выбивается
#  - не выбивается



df.hist(bins = 30, column = '')
plt.show()
print((df[''] > 4).value_counts())
print()

# print(nlargest(3, df['rooms']))
# print()

df[''].info()
print()

# Можно попробовать сделать усы







# 4.4:










pd.set_option('display.max_columns', None)
df.head(15)
```
