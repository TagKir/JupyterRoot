```python
import pandas as pd
import matplotlib.pyplot as plt
import datetime as dt
import numpy as np
import re


# 1:
df1 = pd.read_csv(r"C:\Users\tagir\Downloads\mkrf_shows.csv")
df2 = pd.read_csv(r"C:\Users\tagir\Downloads\mkrf_movies.csv")

# df = df1.join(df2, lsuffix='_caller', rsuffix='_other')

pd.set_option('display.max_rows', None)
pd.options.display.float_format = '{:20,.2f}'.format
df = pd.concat([df1, df2], axis=1)
df = df.replace({np.nan:None})
df['ratings'] = df['ratings'].str.rstrip("%").astype(float)
```

```python
# 2.1 - 2.5
df.loc[1797, 'puNumber'] = '000000000'
df['puNumber'] = df['puNumber'].apply(pd.to_numeric, errors='coerce')
df['show_start_date'] = pd.to_datetime(df['show_start_date'], format='%Y-%m-%dT%H:%M:%S.%f', utc=False)
df['show_start_date'] = pd.to_datetime(df.show_start_date).dt.tz_localize(None)
df['film_studio'] = df['film_studio'].fillna('Неизвестно')
df['director'] = df['director'].fillna('Неизвестно')
df['producer'] = df['producer'].fillna('Неизвестно')
df['age_restriction'] = df['age_restriction'].fillna('Неизвестно')
df['refundable_support'] = df['refundable_support'].fillna(0)
df['nonrefundable_support'] = df['nonrefundable_support'].fillna(0)
df['ofFinSource'] = ((df['refundable_support'] + df['nonrefundable_support']) / df['budget'].fillna(0)).round(4).fillna('Неизвестно')
df['budget'] = df['budget'].fillna('Собственный')
df['financing_source'] = df['financing_source'].fillna('Неизвестно|Сами')
df['ratings'] = df['ratings'].fillna(0)
df.loc[df['ratings'] > 10, 'ratings'] /= 10
df.loc[df['ratings']==0, 'ratings'] = 'Неизвестно'
df['genres'] = df['genres'].fillna('Неизвестно')
df.drop_duplicates(subset=['title', 'show_start_date'], inplace=True, ignore_index=True)


#df['ratings'].value_counts()
#df.query('puNumber == 111000113')
```

```python
# 2.6(1,2,3)
# 1
df['year_of_show_date'] = df['show_start_date'].dt.year
# 2
def FandS_name(df):
    df1 = df.split(' ')
    df2 = re.split(r'[.,]', df )
    if '.' in df:
        return df2[0] + '.' + df2[1]
    elif len(df1) >= 2:
        return df1[0] + ' ' + df1[1]
    else:
        return df1[0]
def Main_genre(df):
    df1 = df.split(',')
    return df1[0]
df['main_director'] = df['director'].apply(FandS_name)
df['main_genre'] = df['genres'].apply(Main_genre)
# 3
# Сделано в 2.1
```
